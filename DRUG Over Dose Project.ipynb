### 1)count occurrences of each indicator for each year

```python
yearly_total_indicators = df.groupby(['YEAR', 'INDICATOR']).size().reset_index(name='TOTAL')
yearly_total = yearly_total_indicators.groupby('YEAR')['TOTAL'].sum()
print(yearly_total)


plt.figure(figsize=(10,6))
yearly_total.plot(kind='bar', color='purple',width=0.6)
plt.xlabel('Year')
plt.ylabel('Total Indicators')
plt.title('Total Indicators per Year')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

```

### 6)What is the trend in drug overdose death rates from 1999 to 2018?

```python

data = {
    'YEAR': [1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018],
    'ESTIMATE': [100, 120, 130, 150, 180, 200, 220, 230, 240, 250, 260, 270, 280, 290, 300, 310, 320, 330, 340, 350]
}
df1 = pd.DataFrame(data)
years = list(range(1999, 2019))
yearly_estimates = []

for year in years:
    total_deaths = df[df['YEAR'] == year]['ESTIMATE'].sum()
    yearly_estimates.append(total_deaths)

plt.figure(figsize=(10, 6))
plt.plot(years, yearly_estimates, marker='o', linestyle='-')
plt.title('Trend of Drug Overdose Death Rates (1999-2018)')
plt.xlabel('Year')
plt.ylabel('Total Drug Overdose Death Rates')
plt.grid(True)
plt.xticks(years, rotation=45) 
plt.tight_layout()
plt.show()
```

### 7)drug overdose death rates vary between males and females over the years?

```python
years_of_interest = [1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]

df = df[df['YEAR'].isin(years_of_interest)]
df_male = df[df['STUB_LABEL'] == 'Male']
df_female = df[df['STUB_LABEL'] == 'Female']

male_yearly = df_male.groupby('YEAR')['ESTIMATE'].mean().reset_index()
female_yearly = df_female.groupby('YEAR')['ESTIMATE'].mean().reset_index()
plt.figure(figsize=(14, 7))

plt.plot(male_yearly['YEAR'], male_yearly['ESTIMATE'], label='Males', marker='o', linestyle='-', color='blue')
plt.plot(female_yearly['YEAR'], female_yearly['ESTIMATE'], label='Females', marker='o', linestyle='-', color='pink')

plt.title('Drug Overdose Death Rates Over Selected Years by Gender')
plt.xlabel('Year')
plt.ylabel('Death Rate (Mean Estimate)') 
plt.legend()
plt.grid(True)
plt.xticks(years_of_interest, rotation=45)
plt.tight_layout()
plt.show()
```

### 8)drug overdose death rates involving heroin differ between males and females?

```python
heroin_data = df[(df['PANEL'] == 'Drug overdose deaths involving heroin') & df['STUB_LABEL'].isin(['Male', 'Female'])]
heroin_deaths_by_gender = heroin_data.groupby('STUB_LABEL')['ESTIMATE'].sum()
plt.figure(figsize=(10, 6))
heroin_deaths_by_gender.plot(kind='bar', color=['blue', 'pink'], width=0.6)
plt.xlabel('Gender')
plt.ylabel('Number of Drug Overdose Deaths Involving Heroin')
plt.title('Drug Overdose Deaths Involving Heroin by Gender')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

### 9)top 5 age group has consistently shown the highest drug overdose death rates?


```python
drug_overdose_data = df[df['INDICATOR'] == 'Drug overdose death rates']

average_death_rates_by_age = drug_overdose_data.groupby('STUB_LABEL')['ESTIMATE'].sum()
top_5_age_groups = average_death_rates_by_age.sort_values(ascending=False).head(5)

print("Top 5 age groups with the highest average drug overdose death rates:")
print(top_5_age_groups)

colors = ['pink', 'lightgreen', 'salmon', 'gold', 'lightcoral', 'lightblue']
plt.figure(figsize=(10, 6))
top_5_age_groups.plot(kind='bar', color=colors)
plt.xlabel('Age Group')
plt.ylabel('Average Death Rates')
plt.title('Top 5 age groups with the highest average drug overdose death rates')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

```

### 10)Distribution of Estimates Across Different Units of Measurement

```python
estimates_by_unit = df.groupby('UNIT')['ESTIMATE'].sum()
print(estimates_by_unit)

plt.figure(figsize=(12, 8))
plt.bar(estimates_by_unit.index,estimates_by_unit.values,width=0.4,color=['salmon', 'gold'],label=estimates_by_unit.index)
plt.title('Distribution of Estimates Across Different Units of Measurement')
plt.xlabel('Unit of Measurement')
plt.ylabel('Estimate')
plt.legend(loc='best')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### 11)What are the top indicators by estimate, and how do they differ across different panels?

```python
top_indicators = df.groupby(['INDICATOR', 'PANEL'])['ESTIMATE'].sum().reset_index()
top_indicators = top_indicators.sort_values(by='ESTIMATE', ascending=False).head(10)
colors = ['green', 'lightgreen', 'salmon', 'gold', 'lightcoral', 'lightblue']
plt.figure(figsize=(10, 8))
plt.barh(top_indicators['INDICATOR'] + ' (' + top_indicators['PANEL'] + ')', top_indicators['ESTIMATE'], color=colors)
plt.xlabel('Sum of Estimates')
plt.ylabel('Indicator (Panel)')
plt.title('Top Indicators by Estimate Across Different Panels')
plt.grid(axis='x')
plt.show()
```

### 12)find out Drug Overdose Deaths by Age Group

```python
overdose_deaths_by_age = df.groupby('AGE')['ESTIMATE'].sum().reset_index()
overdose_deaths_by_age = overdose_deaths_by_age.sort_values(by='ESTIMATE', ascending=False)
print(overdose_deaths_by_age)
colors = ['skyblue', 'lightgreen', 'salmon', 'gold', 'lightcoral', 'lightblue']
plt.figure(figsize=(10, 6))
plt.bar(overdose_deaths_by_age['AGE'], overdose_deaths_by_age['ESTIMATE'], color=colors)
plt.title('Distribution of Drug Overdose Deaths by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
